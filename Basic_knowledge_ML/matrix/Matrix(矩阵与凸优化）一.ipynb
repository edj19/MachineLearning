## 向量与向量空间

### 向量、向量空间、线性组合、基

<font color=blue size=5>向量与向量空间</font>

* 向量加减法，向量与标量数乘，满足交换律，结合律和分配律等
* 向量空间，零向量，单位向量
* 实数向量空间$R^n$

<font color=blue size=5>线性组合与线性无关</font>

线性组合

向量$\textbf{v}_1,\textbf{v}_2,...,\textbf{v}_p\in \textbf{R^n}$，标量$c_1,c_2,...,c_p$，并且向量$\textbf{y}\in \textbf{R^n}$可以表达为
$$
\textbf{y} = c_1\textbf{v}_1+c_2\textbf{v}_2+...+c_p\textbf{v}_p
$$
称为$\textbf{v}_1,\textbf{v}_2,...,\textbf{v}_p\$的线性组合，其中标量$c_1,c_2,...,c_p$称为线性组合的系数或权.


线性无关

向量空间的一组元素称为线性无关，如果其中没有向量可以表示成有限个其他向量的线性组合，反之称为线性相关。
$$
c_1\textbf{v}_1+c_2\textbf{v}_2+...+c_p\textbf{v}_p = 0 \quad
if  \quad c_1=c_2=...=c_p
$$


<font color=blue size=5>线性空间的基</font>

基

* 基是线性空间里的一组向量，使得任何一个向量都可以唯一的表示成这组基的线性组合
* 基向量之间线性无关
* 基给出了定量描述线性结构的方法（坐标系）
* 线性代数的基本问题就是基组的选择问题

<font color=blue size=5>线性生成空间与空间维度</font>

线性生成空间

向量$\textbf{v}_1,\textbf{v}_2,...\textbf{v}_p \in R^n$，它们所有的线性组合的集合称为向量$\textbf{v}_1,\textbf{v}_2,...\textbf{v}_p$的线性生成空间，表示为$Span\{\textbf{v}_1,\textbf{v}_2,...\textbf{v}_p\}$
$$
Span\{\textbf{v}_1,\textbf{v}_2,...\textbf{v}_p\} = {y|\exists c_1,...,\exists c_p ,y=c_1\textbf{v}_1+c_2\textbf{v}_2+...+c_p\textbf{v}_p}
$$

* $\forall i,\textbf{v}_i\in Span\{ \textbf{v}_1,\textbf{v}_2,...\textbf{v}_p\}$
* $0\in Span\{\textbf{v}_1,\textbf{v}_2,...\textbf{v}_p\}$


空间维度

* 线性空间：$Span\{basis\}$基组的线性生成空间
* 空间维度：基的数量
